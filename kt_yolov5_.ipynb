```python
%cd /
from glob import glob
#모든이미지 불러오기
img_list = glob('/content/dataset/export/images/*.jpeg')
#이미지 몇장인지 출력
print(len(img_list))
```

```python
from sklearn.model_selection import train_test_split
#train_test_split() ->training set과 test set을 나눈다
train_img_list, val_img_list = train_test_split(img_list, test_size=0.2, random_state=2000)
#트레이닝 set 80% 밸리데이션 set은 20%로나눔
print(len(train_img_list), len(val_img_list))
```

```python
#training 이미지 경로 ,밸리데이션 이미지 경로를 txt파일로 저장
#training, 밸리데이션 이미지경로를 나눔
# 'w' -> 쓰기지정 리스트형태이기에 join을 씀
with open('/content/dataset/train.txt', 'w') as f:
  f.write('\n'.join(train_img_list) + '\n')

with open('/content/dataset/val.txt', 'w') as f:
  f.write('\n'.join(val_img_list) + '\n')
```

```python
import yaml
#yaml파일 수정하기
#'train'에다가 train.txt 넣고 'val'에다가 val.txt넣기
with open('/content/dataset/bollarddata.yaml', 'r') as f:
  data = yaml.load(f,Loader=yaml.FullLoader)

print(data)

data['train'] = '/content/dataset/train.txt'
data['val'] = '/content/dataset/val.txt'

with open('/content/dataset/bollarddata.yaml', 'w') as f:
  yaml.dump(data, f)

print(data)
```
