```python
%cd /
from glob import glob
#ëª¨ë“ ì´ë¯¸ì§€ ë¶ˆëŸ¬ì˜¤ê¸°
img_list = glob('/content/dataset/export/images/*.jpeg')
#ì´ë¯¸ì§€ ëª‡ì¥ì¸ì§€ ì¶œë ¥
print(len(img_list))
```

```python
from sklearn.model_selection import train_test_split
#train_test_split() ->training setê³¼ test setì„ ë‚˜ëˆˆë‹¤
train_img_list, val_img_list = train_test_split(img_list, test_size=0.2, random_state=2000)
#íŠ¸ë ˆì´ë‹ set 80% ë°¸ë¦¬ë°ì´ì…˜ setì€ 20%ë¡œë‚˜ëˆ”
print(len(train_img_list), len(val_img_list))
```

```python
#training ì´ë¯¸ì§€ ê²½ë¡œ ,ë°¸ë¦¬ë°ì´ì…˜ ì´ë¯¸ì§€ ê²½ë¡œë¥¼ txtíŒŒì¼ë¡œ ì €ì¥
#training, ë°¸ë¦¬ë°ì´ì…˜ ì´ë¯¸ì§€ê²½ë¡œë¥¼ ë‚˜ëˆ”
# 'w' -> ì“°ê¸°ì§€ì • ë¦¬ìŠ¤íŠ¸í˜•íƒœì´ê¸°ì— joinì„ ì”€
with open('/content/dataset/train.txt', 'w') as f:
  f.write('\n'.join(train_img_list) + '\n')

with open('/content/dataset/val.txt', 'w') as f:
  f.write('\n'.join(val_img_list) + '\n')
```
